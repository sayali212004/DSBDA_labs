```python
# Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(y=df['CO(GT)'])
plt.title("Box Plot of CO(GT)")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(6,4))
plt.hist(df['NO2(GT)'], bins=20, color='skyblue')
plt.title("Histogram of NO2(GT)")
plt.xlabel("NO2(GT)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Single Line Graph
plt.figure(figsize=(8,4))
plt.plot(df['C6H6(GT)'], label='C6H6(GT)', color='green')
plt.title("Single Line Graph: C6H6(GT)")
plt.xlabel("Index")
plt.ylabel("C6H6(GT)")
plt.legend()
plt.show()
```

```python
# Multiple Line Graph
plt.figure(figsize=(10,5))
plt.plot(df['CO(GT)'], label='CO(GT)')
plt.plot(df['NOx(GT)'], label='NOx(GT)')
plt.plot(df['NO2(GT)'], label='NO2(GT)')
plt.title("Multiple Line Graph")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
from sklearn import linear_model, metrics
```

```python
X=df[['age']]
Y=df[['thal']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=1)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
ds.shape
```

```python
reg = linear_model.LinearRegression()
```

```python
print(X_train)
```

```python
model = reg.fit(X_train, Y_train)
```

```python
r_sq = reg.score(X_train, Y_train)
```

```python
print("determination coeficient: ", r_sq)
```

```python
print("Intercept: ", model.intercept_)
```

```python
print("slope: ",model.coef_)
```

```python
Y_pred = model.predict(X_test)
print("Predicted Responce: ",Y_pred, sep="\n")
```

```python
plt.scatter(X_train, Y_train, color="black")
plt.plot(X_train, reg.predict(X_train), color="blue", linewidth=3)
plt.title("Age vs Thal")
plt.xlabel("Age")
plt.ylabel("thal")
plt.show()
```

```python
plt.scatter(X_test, Y_test, color="red")
plt.plot(X_test, reg.predict(X_test), color="black", linewidth=3)
plt.title("Age vs Thal")
plt.xlabel("Age")
plt.ylabel("thal")
plt.show()
```
